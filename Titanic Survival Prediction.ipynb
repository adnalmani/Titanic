```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df['Age'].fillna(df['Age'].median(), inplace=True)

df.drop('Cabin', axis=1, inplace=True)

df.isnull().sum()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()

sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, kde=True, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, kde=True, color='red', label='Did not Survive')
plt.title('Age vs. Survival')
plt.legend()
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=df, kind='count')
plt.show()

sns.boxplot(x='Pclass', y='Age', hue='Survived', data=df)
plt.title('Age, Passenger Class, and Survival')
plt.show()
```
